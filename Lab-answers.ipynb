    A. The numbers he/she obtains
    B. The sum of the rolls
    C. Take only the maximum value

Describe the following events:

    For case A: both number is greater than 5
    Case B: The sum is even
    Case C: the maximum is the value of both rolls



omega = {1,2,3,..., until dice max value, if 6 faced dice, 6}
structure = { (1,2), (2, 4), etc., (x, y)}

A. X(x, y) = (x, y)
B. X(x, y) = sum(x, y)
C. X(x, y) = max(x, y)

For case A: the number is greater than 5. 
{X > 5} = (any number greater than 5, any number greater than 5). if 6 faced dice, (6, 6)

Case C: the maximum is the value of both rolls
{X == max(x, y)} = max(any number rolled by dice, any number rolled by dice). if 6 faced dice, max(2, 4)

Describe the measure space and the random variable for:

    Player A wins (no tie).
